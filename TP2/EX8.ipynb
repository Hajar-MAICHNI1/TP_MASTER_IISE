**Exercice 8 : Classe "Personne" avec Gestion d'Amis**
- Modifiez la classe "Personne" pour inclure une liste d'amis.
- Ajoutez des méthodes pour :
 - "ajouter_ami(ami)": pour ajouter un ami à la liste.
 - "afficher_amis()": pour afficher la liste des amis.

```python
class Personne:
    def __init__(self, nom,prenom, age):
        self.nom = nom
        self.age = age
        self.prenom = prenom
        self.amis = []

    def ajouter_ami(self, ami):
        if ami not in self.amis:
            self.amis.append(ami)
            print(f"{ami} a été ajouté à la liste d'amis de {self.nom} {self.prenom}.")
        else:
            print(f"{ami} est déjà un ami de {self.nom} {self.prenom}.")

    def afficher_amis(self):
        if not self.amis:
            print(f"{self.nom} {self.prenom} n'a pas d'amis.")
        else:
            print(f"Les amis de {self.nom} {self.prenom} sont :")
            for ami in self.amis:
                print(ami)

personne1 = Personne("Hajar","MAICHNI", 21)
personne2 = Personne("Kaoula","DAMANI", 22)


personne1.ajouter_ami("Kaoula")
personne1.ajouter_ami("Lamyae")
personne2.ajouter_ami("Hajar")

personne1.afficher_amis()
personne2.afficher_amis()
```
