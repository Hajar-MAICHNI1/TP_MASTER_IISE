**Exercice 2 : Classe "Voiture"**
- Créez une classe "Voiture" qui a les attributs "marque", "modèle", et "année".
- Ajoutez une méthode "afficher_info" qui affiche les informations de la voiture.
- Créez plusieurs instances de "Voiture" et affichez leurs informations.

```python
class Voiture:
  def __init__(self,marque,modele,annee):
    self.marque=marque
    self.modele=modele
    self.annee=annee

  def afficher_info(self):
    print(f"marque: {self.marque}, modele: {self.modele}, annee: {self.annee}")

V1=Voiture("Skoda","Octavia",2019)
V2=Voiture("Ford","Mustang",2020)

V1.afficher_info()
V2.afficher_info()
```
