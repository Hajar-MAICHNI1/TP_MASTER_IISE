**Exercice 7 : Gestion de Bibliothèque**
- Créez une classe "Livre" avec les attributs "titre", "auteur", et "annee_publication".
- Créez une classe "Bibliotheque" qui contient une liste de livres. Ajoutez des méthodes pour :
 - "ajouter_livre(livre)": pour ajouter un livre à la bibliothèque.
 - "afficher_livres()": pour afficher tous les livres de la bibliothèque.

```python
class Livre:
    def __init__(self, titre, auteur, annee_publication):
        self.titre = titre
        self.auteur = auteur
        self.annee_publication = annee_publication

    def __str__(self):
        return f"{self.titre}, écrit par {self.auteur} (Publié en {self.annee_publication})"

class Bibliotheque:
    def __init__(self):
        self.livres = []

    def ajouter_livre(self, livre):
        self.livres.append(livre)

    def afficher_livres(self):
        if not self.livres:
            print("La bibliothèque est vide.")
        else:
            for livre in self.livres:
                print(livre)


livre1 = Livre("Le Petit Prince", "Antoine de Saint-Exupéry", 1943)
livre2 = Livre("Les Misérables", "Victor Hugo", 1862)

bibliotheque1 = Bibliotheque()


bibliotheque1.ajouter_livre(livre1)
bibliotheque1.ajouter_livre(livre2)


bibliotheque1.afficher_livres()
```
